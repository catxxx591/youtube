```python
url1 = 'https://www.youtube.com/watch?v=U87NYhOL8as&list=PLhXu26RzZZTwus4cNbPTcgXXH6oavT6EB&index=1'
res1 = requests.get(url1)
soup = BeautifulSoup(res1.text,'html5lib')
```
